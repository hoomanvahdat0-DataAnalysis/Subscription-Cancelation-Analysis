```c++
with cancel_subs as(
select
   subscription_id,
   cancelation_reason1 as cancelation_reason
from
   public.cancelations

union   

select
   subscription_id,
   cancelation_reason2 as cancelation_reason
from
   public.cancelations

union   

select
   subscription_id,
   cancelation_reason3 as cancelation_reason
from
   public.cancelations

)
select
   cancelation_reason,
   count(subscription_id) as num_subs
from
   cancel_subs 
group by
   1     
```

```c++
with yearly as (
select
   date_trunc('year', cancel_date::date) as cancel_year,
   cancelation_reason,
   count(*) as num_reason
from
   junk.all_cancelation_reasons_jess
group by
   1,2   
)
select
   cancel_year,
   cancelation_reason,
   num_reason,
   sum(num_reason) over(partition by cancel_year) as year_total,
   round(num_reason * 100 / year_total, 2) as perc_reason_annual
from
   yearly         
```
